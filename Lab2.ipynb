### Default Arguments

````python
def volume_cone(height, radius=2.4):
    pi = 3.14
    return pi * radius ** 2 * height / 3
````

````python
# Case1: radius is omitted in volume_cone function call.
v = volume_cone(height=9) # height is passed in by name
print('The volume of a cone is: {}' .format(v))
````

````python
# Case2: the default value of radius=2.4 is overwritten.
v = volume_cone(9, 3) # height and radius are passed in by position
print('The volume of a cone is: {}' .format(v))
````

**c.** Calculate the number of upper case letters, lower case letters and punctuation marks used in a given sentence.

_Note_: A set of built-in methods that you can use on strings are `str.upper()`, `str.isupper()` ,`str.lower()`, `str.islower()`, `str.title()`.

````python
def compute_characters(sentence):
    ### Your implementation here:
    upper = 0
    lower = 0
    punctMarks = 0
    
    """DO STUFF"""
    for character in sentence:
        if character.isupper():
            upper = upper + 1
        if character.islower():
            lower = lower + 1
        if not character.isalpha() and character != ' ':
            punctMarks = punctMarks + 1
    
    ###
    
    print("UPPER CASE {}, lower case {} and punctuation marks {}." .format(upper, lower, punctMarks))

my_str = 'HelLO world'
compute_characters('The AI journey is the perfect opportunity to expand one`s horizons.')
print(my_str.title())
````

### Multiple Arguments `(*args)` and keyword arguments `(**kwargs)`

_Note:_ Used to make the function flexible.

````python
def compute_sum(a, b):
    return a + b
print(compute_sum(4, 5))
````

````python
def compute_sum(*args):
    print('Type of iterable object args {}.' .format(type(args)))
    count = 0
    for a in args:
        count += a
    return count
````

````python
result = compute_sum(0.5, 2.25, 7)
print('Result: ', result)
````

### Task 3

The following formula computes a _Z score_ and measures how far a single raw data value is from the population mean.

\begin{equation*}
z = \frac{X - \mu }{\sigma }
\end{equation*}

where:
* **_X_** is a single raw data value
* `mu` is the population mean
* `sigma` is the population standard deviation

To find the standard deviation, the equation below comes in hand:

\begin{equation*}
\sigma = \sqrt{\frac{\sum \left | X - \mu \right |^{2}}{N}}
\end{equation*}

where **_N_** is the number of data points in the population.

**a.** Using `sum()` and `list comprehension`, compute the mean and the standard deviation for the population defined below:

````python
data =  [4.5, 5, 5.5, 6, 6.25, 7, 15.25, 18, 18.45, 21, 21.45, 23]
print(data)
````

````python
# Your implementation here:
import math
mean = sum(data) / len(data)
print(mean)
std = math.sqrt(sum((x - mean) ** 2 for x in data) / len(data))
print(std)
````

**b.** Define the `z_score()` function and implement the mathematical expression. The obtained values should be stored in a _z score_ values list and rounded to 3 decimals.

````python
# Your implementatio here:

'TO DO'
def z_score(X):
    return (X - mean) / std

print(z_score(18))
print(z_score(4.5))
````

**c.** Add the corresponding elongation of each raw data value into a dictionary.

````python
# Your implementatio here:

'TO DO'
keys = data
values = [z_score(x) for x in keys]

valueToZscoreDict = dict(zip(keys, values))
print(valueToZscoreDict)
````
